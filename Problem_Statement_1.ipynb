```python
def sal_class(a):
    if a==' <=50K':
        b=0
    else:
        b=1
    return b
```

```python
data['Salary_class']=data['salary'].apply(sal_class)
```

```python
data.head()
```

```python
X=data[['age','workclass','education-num','occupation','sex','hours-per-week','country']]
y=data[['Salary_class']]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
work_enco=OneHotEncoder(sparse=False)
work_enco.fit(X[['workclass']])
work_encode=work_enco.transform(X[['workclass']])
work_encode=pd.DataFrame(data=work_encode,columns=work_enco.categories_[0],dtype=int)
work_encode.head()
```

```python
occupation_enco=OneHotEncoder(sparse=False)
occupation_enco.fit(X[['occupation']])
occupation_encode=occupation_enco.transform(X[['occupation']])
occupation_encode=pd.DataFrame(data=occupation_encode,columns=occupation_enco.categories_[0],dtype=int)
occupation_encode.head()
```

```python
sex_enco=OneHotEncoder(sparse=False)
sex_enco.fit(X[['sex']])
sex_encode=sex_enco.transform(X[['sex']])
sex_encode=pd.DataFrame(data=sex_encode,columns=sex_enco.categories_[0],dtype=int)
sex_encode.head()
```

```python
country_enco=OneHotEncoder(sparse=False)
country_enco.fit(X[['country']])
country_encode=country_enco.transform(X[['country']])
country_encode=pd.DataFrame(data=country_encode,columns=country_enco.categories_[0],dtype=int)
country_encode.head()
```

```python
country_encode.shape
```

```python
X=pd.concat([X,work_encode,occupation_encode,sex_encode,country_encode],axis=1)
```

```python
X.head()
```

```python
X=X.drop(['workclass','occupation','sex','country'],axis=1)
X.head()
```

# Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=35)
```

```python
X_train.shape, X_test.shape
```

# Logistic Model

```python
from sklearn.linear_model import LogisticRegression 
```

```python
model=LogisticRegression(random_state=35)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
